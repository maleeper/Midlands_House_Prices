```python
"""
The primary purpose of this script is to facilitate spatial joins between property transaction data and geographic areas.
This enables analysis of housing prices and trends at different geographic levels within the Midlands.

"""
import pandas as pd

# Load the large PCD zipped CSV file with proper encoding
df = pd.read_csv('../data/raw/PCD_OA_LSOA_MSOA_LAD_MAY22_UK_LU.zip', compression='zip', encoding='latin1')


```

```python
# Join with Price Paid Data (PPD) on the `postcode` field

# First load the previously cleaned Price Paid Data (PPD)
ppd = pd.read_csv('../data/clean/pp-monthly-cleaned.csv')
```

```python
ppd.head()
```

```python
# Perform the join
merged_df = pd.merge(ppd, df[['pcds', 'lsoa11cd', 'msoa11nm', 'ladnm']], left_on='postcode', right_on='pcds', how='left')
# The `pcds` column can now be dropped 
merged_df.drop(columns=['pcds'], inplace=True)
# Save the merged dataframe to a new CSV file
merged_df.to_csv('../data/clean/ppd_with_geography.csv', index=False)
```
